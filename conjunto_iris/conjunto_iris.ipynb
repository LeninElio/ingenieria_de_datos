```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print("Tamaño del conjunto de entrenamiento:", len(X_train))
```

```python
print("Tamaño del conjunto de prueba:", len(X_test))
```

```python
clf = SVC(gamma='auto')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Precisión del modelo:", accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
```

```python
print("Precisión:", precision)
print("Recall:", recall)
print("Valor F1:", f1)
```

```python
def predict_species(sepal_length: float, sepal_width: float, petal_length: float, petal_width: float) -> str:
    example = [[sepal_length, sepal_width, petal_length, petal_width]]
    prediction = clf.predict(example)
    species = iris.target_names[prediction][0]
    return species
```

```python
species = predict_species(6.7, 3.0, 5.4, 2.2)
print("Especie predicha:", species)
```
